```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('health.csv',encoding='latin-1')
df.head()
```

```python
edit_df = df[['Name', 'Date of Admission','Medical Condition']]
edit_df.head()
```

```python
unique_values = df['Medical Condition'].unique()
value_counts = df['Medical Condition'].value_counts()
# Filter the value counts for the specific unique values
filtered_counts = value_counts[unique_values]
# Print the filtered counts
print(filtered_counts)
```

```python
diabetes_df = edit_df[edit_df['Medical Condition'] == 'Obesity']
# Print the filtered DataFrame
diabetes_df.shape
```

```python
diabetes_df.head()
```

```python
diabetes_df.dtypes
```

```python
diabetes_df['Date of Admission'] = pd.to_datetime(diabetes_df['Date of Admission'])
# Extract year and month (combined format)
diabetes_df['Year&month'] = pd.to_datetime(diabetes_df['Date of Admission']).dt.strftime('%Y-%m')
# Print the modified DataFrame
diabetes_df.head()
```

```python
diabetes_df = diabetes_df[['Name', 'Year&month','Medical Condition']]
diabetes_df.head(100)
```

```python
# Reset the index
diabetes_df = diabetes_df.reset_index(drop=True)  # Optional: drop the old index
diabetes_df.head()
```

```python
Last_df=diabetes_df
```

```python
# Extract year and month (combined format)
Last_df['Year&month'] = pd.to_datetime(Last_df['Year&month']).dt.strftime('%Y-%m')


# Group by Year&month and count occurrences within each Name group
Last_df = Last_df.groupby(['Year&month', 'Name'])['Name'].count().unstack(fill_value=0)


# Add 'Count' column with total count for each Year&month
Last_df['Count'] = Last_df.sum(axis=1)
Last_df.head(100)
```

```python
Last_df = Last_df.reset_index()
```

```python


# Select only the desired columns
Last_df = Last_df[['Year&month', 'Count']]

Last_df.head(2)
```

```python
Last_df.to_csv('Obesity_data.csv', index=False)  # Optional arguments

print("Dataframe saved successfully!")
```
